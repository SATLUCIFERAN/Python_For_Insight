```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'netflix_titles.csv'
netflix_df = pd.read_csv(file_path)
```

```python
# A detective's first look at the content types: Movies vs. TV Shows
print("--- Content Type Distribution ---")
print(netflix_df['type'].value_counts())
```

```python
# Now, a look at the release years: when was this content released?
print("\n--- Release Year Summary ---")
print(netflix_df['release_year'].describe())
```

```python
# And let's check for missing values across the entire dataset
print("\n--- Missing Values Check ---")
print(netflix_df.isnull().sum())
```

```python
# Visualize the distribution of Release Year with a histogram
plt.figure(figsize=(12, 7))
sns.histplot(data=netflix_df, x='release_year', 
             bins=20, kde=True, color='purple')
plt.title('Distribution of Content Release Years on Netflix', 
          fontsize=16, fontweight='bold')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualize the distribution of Content Type with a bar chart

plt.figure(figsize=(8, 6))
sns.countplot(data=netflix_df, x='type', palette='viridis')
plt.title('Distribution of Movies vs. TV Shows on Netflix', 
          fontsize=16, fontweight='bold')
plt.xlabel('Content Type')
plt.ylabel('Number of Titles')
plt.show()
```
