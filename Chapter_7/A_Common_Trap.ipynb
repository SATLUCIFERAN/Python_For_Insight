```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
```

```python
#We'll use a subset of the numerical columns for our correlation analysis.
correlation_data = titanic_df[['Survived', 'Pclass', 'Age',
                               'SibSp', 'Parch', 'Fare']]
```

```python
# We need to drop rows with any missing values before calculating correlation.
correlation_data = correlation_data.dropna()
```

```python
# Now, we calculate the correlation matrix. The 'corr()' method does all the work for us.
correlation_matrix = correlation_data.corr()
```

```python
# Set a beautiful, professional style for our plot.
sns.set_style("white")
```

```python
# We'll use the correlation matrix from our previous heatmap chapter
correlation_data = titanic_df[['Survived', 'Pclass', 'Age',
                               'SibSp', 'Parch', 'Fare']].dropna()
correlation_matrix = correlation_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='jet', fmt=".2f",
            linewidths=.5, linecolor='black')
plt.title("Correlation Heatmap with " \
          "a Non-Perceptually Uniform Colormap ('jet')",
           fontsize=16, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f",
            linewidths=.5, linecolor='black')
plt.title("Correlation Heatmap with " \
          "a Perceptually Uniform Colormap ('viridis')",
          fontsize=16, fontweight='bold')
plt.show()
```
