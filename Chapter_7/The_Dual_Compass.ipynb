```python
import pandas as pd
import matplotlib.pyplot as plt

file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
df_cleaned = titanic_df.dropna(subset=['Age'])
```

```python
class_summary = df_cleaned.groupby('Pclass')[['Age', 'Fare']].mean()
print("--- Class Summary ---")
print(class_summary)
```

```python
# Set a clean plot style.

plt.style.use('seaborn-v0_8-whitegrid')

# Create the figure and main axes

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the average Age on the main axis (ax).

ax.plot(class_summary.index, class_summary['Age'],
        color='dimgray',
        marker='o',
        linestyle='--',
        label='Average Age')
ax.set_ylabel('Average Age (Years)', color='dimgray')
ax.tick_params(axis='y', labelcolor='dimgray') # Match the label color to the line
ax.set_xlabel('Passenger Class', fontsize=12)
ax.set_title("Average Age and Fare by Passenger Class", fontsize=16, fontweight='bold')

# Create a secondary axis (ax2) that shares the x-axis.
ax2 = ax.twinx()

# Plot the average Fare on the secondary axis (ax2).
ax2.plot(class_summary.index, class_summary['Fare'],
         color='saddlebrown',
         marker='s',
         linestyle='-',
         label='Average Fare')
ax2.set_ylabel('Average Fare ($)', color='saddlebrown')
ax2.tick_params(axis='y', labelcolor='saddlebrown') # Match the label color to the line

# Add a combined legend from both axes.

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Adjust x-axis ticks for clarity

ax.set_xticks(class_summary.index)
ax.set_xticklabels(['1st Class', '2nd Class', '3rd Class'])
plt.tight_layout()
plt.show()
```
