```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
df_cleaned = titanic_df.dropna(subset=['Age'])
```

```python
sns.set_style("whitegrid")

sns.pairplot(df_cleaned, vars=['Age', 'Fare', 'SibSp', 'Parch'], 
             hue='Survived', 
             palette={0: 'tomato', 1: 'forestgreen'})

plt.suptitle("Titanic Pair Plot: Relationships Between " \
             "Key Variables by Survival Status", y=1.02, 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
