```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
```

```python
#We'll use a subset of the numerical columns for our correlation analysis.
correlation_data = titanic_df[['Survived', 'Pclass', 'Age',
                               'SibSp', 'Parch', 'Fare']]
```

```python
# We need to drop rows with any missing values before calculating correlation.
correlation_data = correlation_data.dropna()
```

```python
# Now, we calculate the correlation matrix. The 'corr()' method does all the work for us.
correlation_matrix = correlation_data.corr()
```

```python
# Let's peek at our correlation matrix to understand the numbers
print("--- Correlation Matrix ---")
print(correlation_matrix)
```

```python
# Set a beautiful, professional style for our plot.
sns.set_style("white")
```

```python
# Create a figure to hold our plot
plt.figure(figsize=(10, 8))

# The main heatmap command.
# 'cmap' sets the color scheme, 'annot' puts the numbers on the plot.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            fmt=".2f", linewidths=.5, linecolor='black')

# Add a descriptive title to our plot.

plt.title("Correlation Heatmap of Titanic Variables",
           fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

```
