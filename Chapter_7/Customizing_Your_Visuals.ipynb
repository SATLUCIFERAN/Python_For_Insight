```python
import matplotlib.pyplot as plt
import pandas as pd

file_folder = ('C:/Users/ASUS/Desktop/python_book5/Chapter_7/train.csv')
titanic_df = pd.read_csv(file_folder)

print("--- DataFrame Descriptive Stats for Age and Fare ---")
print(titanic_df[['Age', 'Fare']].describe())
print("\n")
```

```python
df_by_passenger = titanic_df.sort_values(by='PassengerId')\
                            .reset_index(drop=True)
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))
color_age = 'darkslategray'
ax1.set_xlabel('Passenger Index')
ax1.set_ylabel('Age (Years)', color=color_age)
ax1.plot(df_by_passenger.index, df_by_passenger['Age'], 
         color=color_age, linestyle='--', label='Age')
ax1.tick_params(axis='y', labelcolor=color_age)

ax2 = ax1.twinx()  
color_fare = 'saddlebrown'
ax2.set_ylabel('Fare ($)', color=color_fare)  
ax2.plot(df_by_passenger.index, df_by_passenger['Fare'], 
         color=color_fare, linestyle='-', label='Fare')
ax2.tick_params(axis='y', labelcolor=color_fare)

plt.title('Age and Fare Distribution Across Titanic Passengers',
          fontsize=18, fontweight='bold')
ax1.grid(True, linestyle=':', alpha=0.5)

lines = ax1.get_lines() + ax2.get_lines()
ax1.legend(lines, [line.get_label() for line in lines],
           loc='upper right', fontsize=10)

plt.show()
```
