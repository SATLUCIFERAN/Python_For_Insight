```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
```

```python
df_age_clean = titanic_df.dropna(subset=['Age'])
```

```python
sns.set_style("whitegrid")

sns.histplot(data=df_age_clean, x='Age', bins=20, kde=True,
             color='#4e79a7', edgecolor='k')

plt.title("Age Distribution of Titanic Passengers with Density Curve", 
          fontsize=16, fontweight='bold')
plt.xlabel("Age (Years)", fontsize=12)
plt.ylabel("Count / Density", fontsize=12)

average_age = df_age_clean['Age'].mean()
plt.axvline(average_age, color='red', linestyle='--', 
            linewidth=2, label=f'Average Age: {average_age:.2f}')
plt.legend()

plt.tight_layout()
plt.show()
```
