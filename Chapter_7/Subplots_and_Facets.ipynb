```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
```

```python
# We'll drop missing 'Age' values to ensure our plots are clean

titanic_df.dropna(subset=['Age'], inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic_df, x='Age', hue='Pclass',
            fill=True, alpha=0.5, common_norm=False)
plt.title('Age Distribution by Passenger Class (Overlaid)',
           fontsize=16, fontweight='bold')
plt.xlabel('Age (Years)')
plt.show()

# Create a FacetGrid, with a column for each passenger class
g = sns.FacetGrid(titanic_df, col='Pclass', hue='Pclass', height=4, aspect=1.2)

# Map our plotting function (kdeplot) to the grid
g.map(sns.kdeplot, 'Age', fill=True, alpha=0.6)

# Add a main title for the entire figure

plt.subplots_adjust(top=0.85)
g.fig.suptitle("Age Distribution by Passenger Class (Faceted)",
                fontsize=16, fontweight='bold')
plt.show()
```
