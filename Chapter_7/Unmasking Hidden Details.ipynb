```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
file_folder = ('train.csv')
titanic_df = pd.read_csv(file_folder)
```

```python
#Let's drop missing 'Fare' values to be safe
fare_data = titanic_df.dropna(subset=['Fare'])
```

```python
# Graph plot

plt.figure(figsize=(10, 6))
sns.histplot(data=fare_data, x='Fare', bins=50)
plt.title('Distribution of Fares (Standard Linear Scale)',
           fontsize=16, fontweight='bold')
plt.xlabel('Fare ($)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#The same plot, but with a logarithmic y-axis

plt.figure(figsize=(10, 6))
sns.histplot(data=fare_data, x='Fare', bins=50)
plt.yscale('log')
plt.title('Distribution of Fares (Logarithmic Y-Axis)',
           fontsize=16, fontweight='bold')
plt.xlabel('Fare ($)')
plt.ylabel('Number of Passengers (Log Scale)')
plt.show()
```
