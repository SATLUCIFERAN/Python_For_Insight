```python
import matplotlib.pyplot as plt
import pandas as pd

file_folder = ('C:/Users/ASUS/Desktop/python_book5/Chapter_7/train.csv')
titanic_df = pd.read_csv(file_folder)

#The crucial first step: Inspecting the data
print("--- DataFrame Head ---")
print(titanic_df.head())
print("\n--- DataFrame Info ---")
titanic_df.info()
print("\n")

# Sort the DataFrame by 'Fare' to get a clearer progression for the line plot
df_sorted_fare = titanic_df.sort_values(by='Fare').reset_index(drop=True)


# 1. Create a figure—think of this as our blank space
plt.figure(figsize=(10, 6))

# 2. Use the .plot() method to draw our line chart
plt.plot(df_sorted_fare.index, df_sorted_fare['Fare'], 
         marker='.',          # Add a small dot marker at each data point
         linestyle='-',       # Use a solid line to connect the points
         color='steelblue',   # Give the line a steel blue color
         alpha=0.7)           # Make the line slightly transparent


# 3. Add context to our drawing with titles and labels
plt.title("Titanic Passenger Fares (Sorted)", fontsize=16, fontweight='bold')
plt.xlabel("Passenger Index (Sorted by Fare)", fontsize=12)
plt.ylabel("Fare ($)", fontsize=12)

# Add a text annotation to highlight the range of fares
plt.text(x=len(df_sorted_fare)*0.05, y=df_sorted_fare['Fare'].max() * 0.9, 
         s=f"Min Fare: ${df_sorted_fare['Fare'].min():.2f}\nMax Fare: ${df_sorted_fare['Fare'].max():.2f}",
         fontsize=10, bbox=dict(facecolor='white', alpha=0.6, edgecolor='gray', boxstyle='round,pad=0.5'))

# 4. Make sure our grid is visible to help the viewer's eye
plt.grid(True, linestyle='--', alpha=0.6)

# 5. Finally, reveal the drawing to the world

plt.show()
```
