```python
import matplotlib.pyplot as plt
import pandas as pd

file_folder = ('C:/Users/ASUS/Desktop/python_book5/Chapter_7/train.csv')
titanic_df = pd.read_csv(file_folder)

# --- The crucial first step: Inspecting the data ---
print("--- DataFrame Head ---")
print(titanic_df.head())
print("\n--- DataFrame Info ---")
titanic_df.info()
print("\n")
```

```python
pclass_counts = titanic_df['Pclass'].value_counts().sort_index()
pclass_counts
```

```python
pclasses = pclass_counts.index
counts = pclass_counts.values
```

```python
plt.figure(figsize=(8, 6))

plt.bar(pclasses, counts,
        color=['#78a5a3', '#e1b16a', '#c29b7a'], 
        edgecolor='black', 
        linewidth=1)


plt.title("Passenger Count by Class on the Titanic", fontsize=16,fontweight='bold')
plt.xlabel("Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)", fontsize=12)
plt.ylabel("Number of Passengers", fontsize=12)
plt.xticks(pclasses, labels=['1st Class', '2nd Class', '3rd Class'])
plt.grid(axis='y', linestyle=':', alpha=0.7) 


for i, count in enumerate(counts):
    plt.text(pclasses[i], count + 5, str(count), ha='center', 
             fontsize=10, fontweight='bold')

plt.show()
```
